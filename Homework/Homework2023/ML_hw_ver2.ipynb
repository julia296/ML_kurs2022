```python
import pandas as pd
new_data = pd.read_csv("/content/drive/MyDrive/ML_homework/new_data.csv")
```

```python
X = new_data.drop(["Unnamed: 0", "id", "cuisine", "ingredients"], axis = 1)
```

```python
y = new_data.cuisine
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, train_size = 0.75)
```

```python
del new_data
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(n_jobs = -1)

model.fit(X_train, y_train)
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score

accuracy_score(y_test, y_pred)
```

```python
y_pred[:10]
```

```python
from sklearn.tree import DecisionTreeClassifier

model_tree = DecisionTreeClassifier()


model_tree.fit(X_train, y_train)
y_pred_tree = model_tree.predict(X_test)
```

```python
accuracy_score(y_test, y_pred_tree)
```

```python
from sklearn.ensemble import RandomForestClassifier

model_tree = RandomForestClassifier()


model_tree.fit(X_train, y_train)
y_pred_tree = model_tree.predict(X_test)
```

```python
accuracy_score(y_test, y_pred_tree)
```

```python
# для решающего дерева
from sklearn.model_selection import GridSearchCV
from sklearn.tree import DecisionTreeClassifier
# гиперпараметры, которые мы будем перебирать, чтобы найти наилучшую модель
# тут есть критерий разделения наших данных и глубина дерева
tree_param = {'criterion':['gini','entropy'],
              'max_depth':[4,5,6,7,8,9,10,11,12,15,20,30,40,50,70,90,120,150]}



grid_tree = GridSearchCV(DecisionTreeClassifier(), tree_param)
grid_tree.fit(X_train, y_train)


model_dtree = grid_tree.best_estimator_ # наша лучшая модель
y_pred_dtree = model_dtree.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_pred_dtree) # стало немного хуже - не все параметры потому что перебрали
```
